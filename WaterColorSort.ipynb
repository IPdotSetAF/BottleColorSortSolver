```python
BottleCapacity=4
bottles=[
    [1,1,1,1],
    [2,2],
    [3,2,2,3],
    [4,4,4],
    [4]
]

SolveBottles(bottles)
```

```python
import time

Delay=0.5

def SolveBottles(bottles):
    
    time.sleep(Delay)
    PrintBottles(bottles)
    
    hasMove=True
    while(hasMove):
        
        for i in range(BottleCapacity,0, -1):
            matches= FindMatchingBottles(i,bottles)
            if(len(matches)>0):
                break
        
        matchCount=len(matches)
        
        if(matchCount<=1):
            hasMove=False
        elif(matchCount==2):
            if(CanPour(matches[0],matches[1],bottles)):
                Pour(matches[0],matches[1],bottles)
            elif(CanPour(matches[1],matches[0],bottles)):
                Pour(matches[1],matches[0],bottles)
            
                
        time.sleep(Delay)
        PrintBottles(bottles)
        
    print('No more moves')
    
    
def FindMatchingBottles(count ,bottles):
    matches=[]
    for i in range(0, len(bottles)):
        matches.append(i)
        for j in range(i+1, len(bottles)):
            if(bottles[i][-1] == bottles[j][-1]):
                matches.append(j)
        
        if(len(matches) == count ):
            return matches
        
        matches.clear()
        
    return matches

def CanPour(fromB,toB,bottles):
    return (BottleCapacity - len(bottles[toB]) >= TopGroupCount(bottles[fromB]))
    
def TopGroupCount(bottle):
    count=1
    for i in range (len(bottle)-2,-1,-1):
        if(bottle[-1]==bottle[i]):
            count+=1
        else:
            break
            
    return count

def Pour(fromB,toB,bottles):
    while(bottles[fromB][-1] == bottles[toB][-1] & len(bottles[toB])<BottleCapacity):
        bottles[toB].append(bottles[fromB].pop())
```

```python
from IPython.display import clear_output

types = ["@","#","$","&","=","+","/","\\",":"]
#types = ["1","2","3","4","5","6","7","8","9"]
BottleWidth=6
UnitHeight=3

def PrintBottles(bottles):
    
    lines=[]
    
    for i in range(0, BottleCapacity):
        for j in range(0, UnitHeight):
            
            lines.append('')
            
            for bottle in bottles:
                if(i < len(bottle)):
                    lines[-1]+=GenerateLayer(bottle[i]-1)
                else :
                    lines[-1]+=GenerateEmpty()
                    
    
    lines.insert(0, f'{len(bottles) * GenerateBase()}')
    lines.append(f'{len(bottles) * GenerateEmpty()}')
    lines.append('')
    
    lines.reverse()
    
    clear_output(wait=True)
    
    for line in lines:
        print(line)
    

def GenerateLayer(index):
    return f'|{types[index] * BottleWidth}|\t'

def GenerateEmpty():
    return f'|{" "* BottleWidth}|\t'

def GenerateBase():
    return f'|{BottleWidth * "_"}|\t'
```
